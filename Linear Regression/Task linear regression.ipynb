```python
import numpy as np 
import matplotlib.pyplot as plt

# Define x and y data points
x = np.array([2, 4, 6, 8, 10, 12])
y = np.array([5, 3, 3, 8, 5, 6])

# Define the hypothesis function
def pred(x, w, b):
    return w * x + b

# Define the cost function
def cost(x, y, w, b):
    return np.sum((pred(x, w, b) - y) ** 2) / (2 * len(x))

# Initialize w and b values
w = 0
b = 0

# Calculate the cost for each case
print("Cost for y1 = x:", cost(x, y, 1, 0))
print("Cost for y2 = 0.5*x:", cost(x, y, 0.5, 0))
print("Cost for y3 = 0.5*x+4:", cost(x, y, 0.5, 4))

# Plot the points and lines
plt.title("Linear Regression")
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x, y)
plt.plot(x, pred(x, 1, 0), label="y1 = x")
plt.plot(x, pred(x, 0.5, 0), label="y2 = 0.5*x")
plt.plot(x, pred(x, 0.5, 4), label="y3 = 0.5*x + 4")
plt.legend()
plt.show()

# Create w and b arrays for 3D scatter plot
w_vals = np.linspace(-2, 2, 30)
b_vals = np.linspace(-2, 2, 30)
w, b = np.meshgrid(w_vals, b_vals)

# Calculate J(w, b) for each w and b value
J_vals = np.zeros((w.shape[0], w.shape[1]))
for i in range(w.shape[0]):
    for j in range(w.shape[1]):
        J_vals[i,j] = cost(x, y, w[i,j], b[i,j])

# Plot the 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(w, b, J_vals)
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('J(w, b)')
plt.show()
```
